Algorithm:
1.	Start
2.	Initialize phone_options, cart, total_price, and membership_points.
3.	While True: 4. Display available phone options.
5.	Get user input for desired smartphone, memory size, and quantity.
6.	Calculate the price for the selected phone.
7.	Add the selected phone to the cart and update the total price.
8.	Ask if the user wants to continue shopping.
9.	If not, break the loop.
4.	Get user input for the payment method.
5.	Apply discount based on the payment method.
6.	Calculate membership points earned based on the total price.
7.	If membership points are greater than or equal to 10, offer to redeem points.
8.	Display final results including the total price, membership points earned, and discounted price if applicable.
9.	End

```python
def smartphone_shopping():
    phone_options = {
        "iPhone 12": {"128": 3199, "256": 3599},
        "iPhone 13 Mini": {"64": 2499, "128": 2799},
        "iPhone 13 Pro": {"256": 4999, "512": 5999},
        "Samsung Galaxy S1": {"128": 2899, "256": 3299},
        "Samsung Galaxy S2": {"64": 1799, "128": 1999}
    }
    cart = []
    total_price = 0
    membership_points = 0

    while True:
        display_phone_options(phone_options)
        model, memory_size, quantity = get_user_input(phone_options)
        if model is None:
            break
        price = calculate_price(phone_options[model][memory_size], quantity)
        cart.append((model, memory_size, quantity, price))
        total_price += price

        if not continue_shopping():
            break

    payment_method = get_payment_method()
    total_price, discounted_price = calculate_discount(payment_method, total_price)
    membership_points = earn_points(total_price)

    if membership_points >= 10:
        total_price, redeemed_discount = redeem_points(total_price, membership_points)
        discounted_price -= redeemed_discount

    display_results(total_price, membership_points, discounted_price)


def display_phone_options(phone_options):
    print("Welcome to the Smartphone Shopping Center!")
    print("Available Options:")
    for idx, (model, sizes) in enumerate(phone_options.items(), start=1):
        print(f"{idx}. {model}")
    print("Enter the number corresponding to your desired smartphone or type 'quit' to finish: ")


def get_user_input(phone_options):
    while True:
        selection = input()
        if selection.lower() == 'quit':
            return None, None, None
        elif selection.isdigit() and int(selection) in range(1, len(phone_options) + 1):
            model = list(phone_options.keys())[int(selection) - 1]
            memory_size = input(f"Enter the memory size (GB) for {model} ({', '.join(phone_options[model].keys())}): ")
            if memory_size not in phone_options[model]:
                print("Invalid memory size. Please try again.")
                continue
            quantity = int(input("Enter the quantity: "))
            return model, memory_size, quantity
        else:
            print("Invalid selection. Please try again.")


def calculate_price(price, quantity):
    return price * quantity


def continue_shopping():
    while True:
        choice = input("Enter the number corresponding to your desired smartphone or type 'quit' to finish: ").lower()
        if choice in ['yes', 'no']:
            return choice == 'yes'
        else:
            print("Invalid choice. Please enter 'yes' or 'no'.")


def get_payment_method():
    print("\nPayment Methods:")
    print("1. FAB Banking Cards")
    print("2. ADIB Banking Cards")
    print("3. Other Banking Cards")
    print("4. Cash")
    while True:
        selection = input("Enter the number corresponding to your payment method: ")
        if selection.isdigit() and int(selection) in range(1, 5):
            return selection
        else:
            print("Invalid selection. Please try again.")


def calculate_discount(payment_method, total_price):
    discounts = {
        "1": 0.1,
        "2": 0.15,
        "3": 0.05,
        "4": 0.0
    }
    discount_rate = discounts.get(payment_method, 0.0)
    discounted_price = total_price * (1 - discount_rate)
    print(f"\nTotal Price: {total_price:.2f} AED (After {discount_rate * 100}% discount)")
    return total_price, discounted_price


def earn_points(total_price):
    return total_price // 100


def redeem_points(total_price, membership_points):
    redemption_rate = 5
    max_discount = total_price * 0.5
    discount = min(membership_points * redemption_rate, max_discount)
    updated_total_price = total_price - discount
    print(f"Membership Points Earned: {membership_points} (1 point for every 100 AED spent)")
    print(f"Discount Applied: {discount:.2f} AED")
    return updated_total_price, discount


def display_results(total_price, membership_points, discounted_price):
    print(f"Updated Total Price: {discounted_price:.2f} AED")
    print("Thank you for shopping with us!")


# Call main function
smartphone_shopping()
```
