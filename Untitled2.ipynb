Introduction
The Library Book Management System is a Python program that allows users to manage a library's
collection of books. The program provides the following functionalities:
1. Display Books: View the list of available books in the library.
2. Borrow a Book: Borrow a book from the library, or request it if it's not available.
3. Return a Book: Return a borrowed book to the library.
The program maintains three lists: library_books (for available books), borrowed_books (for books
currently borrowed), and requested_books (for books that are requested by users but not available).

Code Structure
The code is structured as follows:
1. Import necessary libraries (none in this case).
2. Initialize the library_books list, borrowed_books list, and requested_books list.
3. Define functions for displaying books, borrowing books, and returning books.
4. Implement a while loop for user interaction.
5. Display a menu to allow the user to choose an action.
Functions1. display_books(library_books): This function displays the list of available books in the library.
2. borrow_book(book_id): This function allows the user to borrow a book based on its ID. It checks if the
book is available in the library_books list and updates the lists accordingly.
3. return_book(book_id): This function allows the user to return a borrowed book based on its ID. It
checks if the book is in the borrowed_books list, updates the lists, and increments the number of times
the book has been borrowed

```python
library_books = []
```

```python
def display_books(library_books):

        for book in library_books:
            print(f"Book ID: {book[0]}")
            print(f"Title: {book[1]}")
            print(f"Authors: {book[2]}")
            print(f"Year: {book[3]}\n")
```

```python
def borrow_book(book_id):
    for book in library_books:
        if book[0] == book_id:
            library_books.remove(book)
            borrowed_books.append(book)
            print(f"The book with the ID {book_id} is successfully loaned to you now.")
            return
    for book in borrowed_books:
        if book[0] == book_id:
            requested_books.append(book)
            print(f"The book with the ID {book_id} is not available right now. However, we have requested it for you.")
            return
    print(f"The book with the ID {book_id} is unfortunately not available in the system. Please re-check the ID.")
```

```python
def return_book(book_id):
    for book in borrowed_books:
        if book[0] == book_id:
            borrowed_books.remove(book)
            library_books.append(book)
            if book in requested_books:
                requested_books.remove(book)
            print(f"The book with the ID {book_id} is successfully returned.")
            return
    print(f"The book with the ID {book_id} cannot be returned. Please re-check the ID.")
```

```python
borrowed_books = []
requested_books = []

# Sample book entries
library_books.append(['0001', 'Introduction to Chemistry', ('John Smith'), 2012])
library_books.append(['0002', 'Software Engineering', ('Ian Sommerville'), 2002])
library_books.append(['0003', 'Functional Programming', ('Jack Widman', 'Ahmad Yusif'), 2023])
```

```python
while True:
    print("\nLibrary Book Management System")
    print("1. Display Books")
    print("2. Borrow a Book")
    print("3. Return a Book")
    print("4. Quit")
    choice = input("Enter your choice: ")

    if choice == "1":
        print("\nLibrary Books:")
        display_books(library_books)
    elif choice == "2":
        book_id = input("Enter the book ID you want to borrow: ")
        borrow_book(book_id)
    elif choice == "3":
        book_id = input("Enter the book ID you want to return: ")
        return_book(book_id)
    elif choice == "4":
        break
    else:
        print("Invalid choice. Please select a valid option.")
```
